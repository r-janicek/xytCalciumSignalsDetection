## Load movies, annotations & predictions

```python
# Config model to load
training_name = "TEMP_new_annotated_peaks_physio"
epoch = 100000

preds_dir = os.path.join("trainings_validation", training_name)
```

```python
# Load movies
input_movies = load_movies_ids(
    data_folder=dataset_dir, ids=movie_ids, names_available=True, movie_names="video"
)

# Load annotations
# segmentation masks
ys = load_annotations_ids(
    data_folder=dataset_dir, ids=movie_ids, mask_names="class_label"
)

# event masks
ys_events = load_annotations_ids(
    data_folder=dataset_dir, ids=movie_ids, mask_names="event_label"
)

# Load predictions
sparks_filenames = {
    movie_id: os.path.join(preds_dir, f"{training_name}_{epoch}_{movie_id}_sparks.tif")
    for movie_id in movie_ids
}
puffs_filenames = {
    movie_id: os.path.join(preds_dir, f"{training_name}_{epoch}_{movie_id}_puffs.tif")
    for movie_id in movie_ids
}
waves_filenames = {
    movie_id: os.path.join(preds_dir, f"{training_name}_{epoch}_{movie_id}_waves.tif")
    for movie_id in movie_ids
}

sparks = {
    movie_id: np.asarray(imageio.volread(f)) for movie_id, f in sparks_filenames.items()
}
puffs = {
    movie_id: np.asarray(imageio.volread(f)) for movie_id, f in puffs_filenames.items()
}
waves = {
    movie_id: np.asarray(imageio.volread(f)) for movie_id, f in waves_filenames.items()
}
```

### Compute class weights for numbers of events

To use in order to get frames with the most annotated events in a weighted way

```python
# compute class weights
w_dict = compute_class_weights_instances(
    ys=list(ys.values()), ys_events=list(ys_events.values())
)
```

### Process predictions to obtain segmentation masks with small objects removed

```python
preds_segmentation = {}

for movie_id in movie_ids:
    raw_preds_dict = {
        "sparks": sparks[movie_id],
        "puffs": puffs[movie_id],
        "waves": waves[movie_id],
    }
    _, preds_segmentation[movie_id], _ = process_raw_predictions(
        raw_preds_dict=raw_preds_dict, input_movie=input_movies[movie_id]
    )
```

## Compute statistics

```python
# configure dataframe
cols = [
    "movie ID",
    "frame ID",
    "sparks IoU",
    "puffs IoU",
    "waves IoU",
    "average IoU",
    "sparks events",
    "puffs events",
    "waves events",
    "sparks weighted events",
    "puffs weighted events",
    "waves weighted events",
    "sum weighted events",
]

frames_stats_df = pd.DataFrame(columns=cols)

# iterate over movies and frame to compute statistics
for movie_id in movie_ids:
    # remove frames ignored by loss function
    sample_ys = trim_and_pad_video(
        video=ys[movie_id], n_margin_frames=ignore_frames_loss
    )

    sample_ys_events = trim_and_pad_video(
        video=ys_events[movie_id], n_margin_frames=ignore_frames_loss
    )

    # sample_preds is a dict with keys 'sparks','puffs' and 'waves'
    sample_preds = {}  # where predicted segmentation will be stored
    for event_type in config.classes_dict.keys():
        if event_type in ["background", "ignore"]:
            continue
        sample_preds[event_type] = trim_and_pad_video(
            video=preds_segmentation[movie_id][event_type],
            n_margin_frames=ignore_frames_loss,
        )

    # mask ignored during training
    ignore_mask = sample_ys == config.ignore_index

    for frame_id in range(len(sample_ys)):
        # create frame dict that will be appended to dataframe
        frame_dict = {
            "movie ID": movie_id,
            "frame ID": frame_id,
            "average IoU": 0,
            "sum weighted events": 0,
        }

        for event_type, event_label in config.classes_dict.items():
            if event_type in ["background", "ignore"]:
                continue
            # get binary masks corrisponding to event type
            ys_frame_class = sample_ys[frame_id] == event_label
            preds_frame_class = sample_preds[event_type][frame_id]

            # compute Intersection over Union score
            iou = compute_iou(
                ys_roi=ys_frame_class,
                preds_roi=preds_frame_class,
                ignore_mask=ignore_mask[frame_id],
            )
            frame_dict[event_type + " IoU"] = iou
            frame_dict["average IoU"] += iou

            # get event mask corrisponding to event type
            ys_events_frame_class = np.where(
                ys_frame_class, sample_ys_events[frame_id], 0
            )
            # count number of events in frame
            frame_dict[event_type + " events"] = (
                np.unique(ys_events_frame_class).size - 1
            )
            # count number of weighted events in frame
            frame_dict[event_type + " weighted events"] = (
                w_dict[event_type] * frame_dict[event_type + " events"]
            )
            frame_dict["sum weighted events"] += (
                w_dict[event_type] * frame_dict[event_type + " events"]
            )

        frame_dict["average IoU"] /= len(
            [
                event_type
                for event_type in config.classes_dict.keys()
                if event_type not in ["background", "ignore"]
            ]
        )

        # create df from frame dict
        frame_df = pd.DataFrame([frame_dict])
        frames_stats_df = pd.concat([frames_stats_df, frame_df])
```

### Discard empty frames (if both ys and preds are empty)

Idea: if nb pixels in annotations == 0 and IoU avg == 1 ==> nb pixels in preds = 0

```python
frames_stats_df = frames_stats_df.loc[
    (frames_stats_df["average IoU"] != 1)
    | (frames_stats_df["sum weighted events"] != 0)
]
```
